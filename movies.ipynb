```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv('movies_data.csv')
```

```python
df 
```

```python
df['combined'] = df['Genre'] + ' ' + df['Director'] + ' ' + df['Actor 1'] + ' ' + df['Actor 2'] + ' ' + df['Actor 3']
```

```python
df['combined'] = df['combined'].apply(lambda x: x.lower())
```

```python
df['combined']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
cv = TfidfVectorizer(stop_words='english')
```

```python
feature_matrix = cv.fit_transform(df['combined'])
```

```python
feature_matrix.toarray()
```

```python
scale = MinMaxScaler()
```

```python
scalar = scale.fit_transform(df[['Duration', 'Rating', 'Votes']])
```

```python
import numpy as np
```

```python
movie_vector = np.hstack([feature_matrix.toarray(), scalar])
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
neigh = NearestNeighbors(metric='cosine')
```

```python
neigh.fit(movie_vector)
```

```python
def get_recommendation(movie_name):
    idx = df[df['Name']==movie_name].index[0]
    dist, indices = neigh.kneighbors([movie_vector[idx]], n_neighbors=6)
    recommended_movies = df.iloc[indices[0][1:]]
    return recommended_movies[['Name']]
```

```python
name = '...Aur Pyaar Ho Gaya'
get_recommendation(name)
```
